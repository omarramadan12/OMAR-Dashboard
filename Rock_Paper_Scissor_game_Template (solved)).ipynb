### Description 
> #### In this game, the user gets the first chance to pick the option among Rock, paper and scissors. After that computer select from the remaining two choices(randomly), the winner is decided as per the rules.
- Rock vs paper-> paper wins
- Rock vs scissor-> Rock wins
- paper vs scissor-> scissor wins.

Tasks:

1. Take input from user as R, P, S
2. Use `random` module to choose a random value for computer
3. Decide who is the winner and announce it

```python
# Task 1

def get_choice():
    choice = input("Enter your choice (R for Rock, P for Paper, S for Scissors): ").lower()

    while choice not in ['r', 'p', 's']:
        choice = input("Input is invalid. Please try again: ").lower()

    return choice
```

```python
# Task 2

import random

def get_random_choice():
    choices = ['r', 'p', 's']

    return random.choice(choices)
```

```python
# Task 3

def decide(user_choice, computer_choice):
    print(f"Computer Choice: {computer_choice} - Player Choice: {user_choice}")
    if user_choice == 'r':
        if computer_choice == 'r':
            print("Draw.")
        elif computer_choice == 'p':
            print("Computer Wins.")
        else:
            print("User Wins.")
    elif user_choice == 'p':
        if computer_choice == 'r':
            print("User Wins.")
        elif computer_choice == 'p':
            print("Draw.")
        else:
            print("Computer Wins.")
    else:
        if computer_choice == 'r':
            print("Computer Wins.")
        elif computer_choice == 'p':
            print("User Wins.")
        else:
            print("Draw.")
```

```python
def main():
    playing = True

    while playing:
        user = get_choice()
        computer = get_random_choice()
        decide(user, computer)

        again = input("Play again? (y/n): ")

        if again.lower() == 'n':
            playing = False

    print("Goodbye.")
```

```python
main()
```
